## Characteristics of the Dot Product

### What is the Dot Product?

When multiplying two vectors together, the result is a scalar value. The concise mathematical defintion of this multiplication, called a dot product (also known as an inner product), is:

> Given $v = (v_{1}, v_{2})$ and $w = (w_{1}, w_{2})$, the **inner product** of these vectors is the number $v \cdot w$ resulting from the operation $v \cdot w = v_{1}w_{1} + v_{2}w_{2}$

The underlying requirement for a dot product is that two vectors being multiplied **must** be of the same length.

### Special Results of the Dot Product

Suppose the vectors $v = (4, 2)$ and $w = (-1, 2)$ are given. The inner product of these two vectors results in a **zero** dot product:

$\left[
\begin{array}{c}
4 \\
2 \\
\end{array}
\right]
\cdot
\left[
\begin{array}{c}
-1 \\
2 \\
\end{array}
\right] = -4 + 4 = 0$ 

Note the result of the inner product of these two vectors is **zero**. What this means, is that the angle between these two vectors are _perpendicular_ (i.e. 90$^{\circ}$).

### Operations 

#### Communicativity of the Dot Product

Given a vector $v$ and vector $w$, the dot product of $v \cdot w$ is equivalent to $w \cdot v$. This illustrates the commutative property of the dot product.

#### Dot Product for $n$ Number of Elements in a Vector

Furthermore, if one were given the vectors $c = (c_{1}, c_{2}, c_{3})$ and $d = (d_{1}, d_{2}, d_{3})$ and asked to find the inner product, it would be analogous to previous 2-element vectors. Finding the inner product would take the form:

$c \cdot d = (c_{1}, c_{2}, c_{3}) \cdot (d_{1}, d_{2}, d_{3}) = c_{1}d_{1} + c_{2}d_{2} + c_{3}d_{3}$

This can be continued and generalized to where for any two given vectors of the same number of elements $n$ for which the inner product is desired to be found, the inner product is calculated thus:

$c \cdot d = (c_{1} \cdots c_{n}) \cdot (d_{1} \cdots d_{n}) = c_{1}d_{1} \ + \cdots + \ c_{n}d_{n}$

## Lengths and Unit Vectors

### What Is the Length of a Vector?

The length of a vector $v$ of $n$-elements is defined as:

> $length = \| v \| = \sqrt{v \cdot v} = (v_{1}^{2} + \cdots + v_{n}^{2})^{\frac{1}{2}})$

The $\|v\|$ denotes the ordinary length of the arrow the represents the vector $v$. The formalized definition is derived from the Pythagorean Theorem $a^{2} + b^{2} = c^{2}$. This is best explained with an example:

#### How Length of a Vector Is Derived

Suppose one is given the vector $v = (1, 2)$. As was given in tutorial 1-1, $v$ can be decomposed into the Cartesian components $x = 1$ and $y = 2$. These two components can be considered as being the sides of a right triangle. To find the third side, one applies Pythagorean's Theorem $1^{2} + 2^{2} = side^{2}$ and solves for the missing side which results as $\sqrt{5}$.

Knowing what the dot product is, to find the length of a particular vector can be directly mapped from the Pythagorean Theorem. Using $v$, the notation of length can be substituted for the missing side:

> $1^{2} + 2^{2} = \|v\|^{2}$

Which is empirically equivalent to the general definition of length. This is further able to be applied when a vector has $n$-elements.

### What is a Unit Vector?

In linear algebra, the word _"unit"_ indicates that the measurement of something equals one. Accordingly, a unit vector is defined as:

> A unit vector $u$ is a vector whose length equals one. (i.e. $u \cdot u = 1$)

#### How To Determine a Unit Vector in the Same Direction of a Non-Zero Vector

For a given nonzero vector, $v$, a corresponding unit vector can be found that follows the same direction as that vector. This formulation is defined as:

> $\vec{u} = \frac{v}{\|v\|}$

**NOTE:** Going forward, these tutorials will use the arrow hat notation to represent unit vectors. This is common notation for educational institutes. (i.e. the unit vector for the general non-zero vector, $u$, will be represented as $\vec{u}$)

### How Do You Find Angle Measurements between Two Vectors

Previously was demonstrated that if one were to calculate the dot product between two vectors and its result was zero, then these two vectors are perpandicular to each other. However, the question remains: what if the dot product is _not_ zero between two vectors?

To answer this question, one can use the Cosine formula.

#### What Is the Cosine Formula? 

The cosine formula finds the cosine angle between two vectors. In actuality, its formulation is the dot product of two distinct unit vectors. Given two unit vectors, $\vec{u}$ and $\vec{v}$, the condition to find the cosine angle between these two vectors is that $| \vec{u} \cdot \vec{v}| \leq 1$. Once this is met, the general cosine formulation can be used:

Given the nonzero vectors $v$ and $w$ then,

> $\frac{v \cdot w}{\|v\| \|w\|} = cos(\theta)$

#### Important Results from the Cosine Formula

One important result from the Cosine Formula is known as the Schwarz Inequality which states that the dot product of the dot product of $\vec{v} \cdot \vec{w}$ never exceeds one. Mathematically, it is stated as:

> $|\boldsymbol{v} \cdot \boldsymbol{w}| \leq\|\boldsymbol{v}\|\|\boldsymbol{w}\|$

Furthermore, the Triangle Inequality follows from the Cosine Formulation as well:

> $\| v + w\| \leq \|v\| + \|w\|$

The Triangle Inequality can be thought of as the the hypotenuse of a triangle is less than or equal to the two remaining sides of that triangle.
