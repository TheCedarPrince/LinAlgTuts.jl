```julia
using Plots
using LaTeXStrings
```

```julia
y(x, x_1, y_1, x_2, y_2) = x .* (y_2 - y_1) ./ (x_2 - x_1) .- x_1 .* (y_2 - y_1) ./ (x_2 - x_1) .+ y_1
```

## What Is the Fundamental Problem of Linear Algebra?

The fundamental problem of linear algebra is to solve a system of linear equations. Linear equations are only ever multiplied by numbers, never by other variables.

## Examining the Row Picture of Linear Systems

If given two equations:

(1) $\ x - 2y = 1$

(2) $\ 3x + 2y = 11$

Examining these equations independently yields two lines converging at one point:

```julia
input = -5:5;
output = y(input, 1, 0, 3, 1);
```

```julia
plot(input,
     output, 
     framestyle=:zerolines,
     label="Equation (1)",
     title="Row Picture of Equations (1) & (2)",
     xlim=(-5, 5),
     ylim=(-5, 5),
     linewidth = 3
     );
```

```julia
input = -5:5;
output = y(input, 0, 5.5, 3, 1);
```

```julia
plot!(input,
     output, 
     label="Equation (2)",
     linewidth = 3
     );
```

```julia
scatter!(
    (3, 1),
    annotations = (3.1, 0.8, Plots.text("Solution", :top, :left)),
    marker = 10,
    label = false,
)
```

The point that lies on both lines, solves both equations simultaneously.

## Examining the Column Picture of Linear Systems

As shown in the previous section, Section 1.3, linear systems of equations can be re-written as vectors. This "vector equation" enables a different view of the equations (1) and (2) such that we now view them as columns:

(3) $\ x\left[\begin{array}{l}1 \\ 3\end{array}\right]+y\left[\begin{array}{r}-2 \\ 2\end{array}\right]=\left[\begin{array}{r}1 \\ 11\end{array}\right]=b$

In essence, viewing these equations in a columnwise vector fashion, still asks the same question. What unique value exist that yield a solution for both equations simultaneously? Unlike the row picture, instead of plotting the rows to find this point, we ask in the column picture, what _combination of these vectors equal the vector on the right of this equation_ solves the given vector equation, (3)?

Solving this problem is relatively straightforward to do. All one would need to do, is plug in values for $x$ and $y$ until a solution is found. The solution here, would be $x = 3$ and $y = 1$. This produces the following visualization:

```julia
input = 0:3;
output = y(input, 0, 0, 3, 9);
```

```julia
plot(input,
     output, 
     arrow = :head,
     label="3 * Column 1",
     linewidth = 3
     );
```

```julia
input = -2:0;
output = y(input, 0, 0, -2, 2);
```

```julia
plot!(input,
     output, 
     arrow = :tail,
     framestyle=:zerolines,
     label="Column 2",
     title="Column Picture of Equations (1) & (2)",
     xlim=(-5, 5),
     ylim=(0, 15),
     linewidth = 3
     );
```

```julia
input = 0:1;
output = y(input, 0, 0, 1, 11);
```

```julia
plot!(input,
     output, 
     arrow = :head,
     label="Solution b",
     linewidth = 3
     )
```

In this plot, the vector shown by $3 * Column \ 1$ corresponds to the vector $(3, 9)$. Accordingly, $Column \ 2$ is $(-2, 2)$ and $Solution \ b$ is the final vector $(1, 11)$. The operations carried out for this problem was first scalar multiplication and then finally, linear combination to come up with this solution:

(4) $\ 3\left[\begin{array}{l}1 \\ 3\end{array}\right]+1\left[\begin{array}{r}-2 \\ 2\end{array}\right]=\left[\begin{array}{r}1 \\ 11\end{array}\right]$

Although the column picture is useful, it is not in a manageable form. A better form, would be to put the column picture into the format of a matrix equation. This would look something like this:

$\left[
\begin{array}{cc}
1 & -2 \\
3 & 2 \\
\end{array}
\right]
\left[
\begin{array}{c}
x \\
y \\
\end{array}
\right] = 
\left[
\begin{array}{c}
1 \\
11 \\
\end{array}
\right]$



If one were to examine the matrix $\begin{equation}
\left[
\begin{array}{cc}
1 & -2 \\
3 & 2 \\
\end{array}
\right]
\end{equation}$, these values represent the coefficients of equations (1) and (2). Accordingly, this particular matrix $A$, is called a **coefficient matrix**. Furthermore, the vector corresponds to the input vector, $x$, of the matrix equation $Ax = b$. Taking the dot product on the left of the equation gives back the form found in (3).
