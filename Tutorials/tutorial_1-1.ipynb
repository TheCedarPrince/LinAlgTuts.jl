## What is a Vector?

A vector can be thought of in two ways:

- **Definition 1:** A vector is a column of a matrix.
- **Definition 2:** A vector is that which has the algebra of a vector space.

These definitions are credited to Jiahao Chen [1] in his talk on _Taking Vector Transposes Seriously_. The first defintion will provide a more direct understanding of vector algebra. For now, operate with this defintion where given some matrix, _A_ (which is a rectangular representation of numbers), a vector, _b_, would be a vertical slice of that matrix:

$A = \left[
\begin{array}{ccc}
a & b & c \\
d & e & f \\
g & h & i \\
\end{array}
\right], \
b = \left[
\begin{array}{c}
a \\
b \\
g \\
\end{array}
\right] $

The second defintion will make more sense and open one to better way of thinking about vectors in the future.

## How Does Linear Combination Operate?

Let $v$ and $w$ be any two vectors and let $\alpha$ and $\beta$ represent any real numbers. Multiplying $v$ and $w$ by $c$ and $d$ respectively yields $cv$ and $dw$.

Adding $cv$ and $dw$ together is the idea of a linear combination.

$\alpha v + \beta w = \alpha\begin{bmatrix} v_{1}\\ v_{2} \end{bmatrix} + \beta\begin{bmatrix} w_{1}\\ w_{2} \end{bmatrix} = \begin{bmatrix} \alpha v_{1} + \beta w_{1}\\ \alpha v_{2} + \beta w_{2} \end{bmatrix}$

## Reading Vector $u$ in the Cartesian Coordinate Plane

In the case of 

$\alpha\left[
\begin{array}{c}
1 \\
1 \\
\end{array}
\right]
+
\beta\left[
\begin{array}{c}
2 \\
3 \\
\end{array}
\right]$

The entire xy plane is filled. They produce every $\left[
\begin{array}{c}
x \\
y \\
\end{array}
\right]$

In three element column vectors, a **plane** is filled in the *xyz* space:

$\alpha\left[
\begin{array}{c}
1 \\
1 \\
1 \\
\end{array}
\right]
+
\beta\left[
\begin{array}{c}
2 \\
3 \\
4 \\
\end{array}
\right]$

and can be analogously understood to produce in three dimensions as $\left[
\begin{array}{c}
x \\
y \\
z \\
\end{array}
\right]$
