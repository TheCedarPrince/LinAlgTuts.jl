# **Tutorial: The Column Space of a Matrix**

## **Column Space of A**

[TBD]

$Ax \ = \ \left[
\begin{array}{ccc}
1 & 4 & 5 \\
3 & 2 & 5 \\
2 & 1 & 3 \\
\end{array}
\right]\
\left[
\begin{array}{c}
x_{1} \\
x_{2} \\
x_{3} \\
\end{array}
\right]\
= \
\left[
\begin{array}{c}
1 \\
3 \\
2 \\
\end{array}
\right]\
x_{1} + \
\left[
\begin{array}{c}
4 \\
2 \\
1 \\
\end{array}
\right]\
x_{2} + \
\left[
\begin{array}{c}
5 \\
5 \\
3 \\
\end{array}
\right]\
x_{3}
$

$Ax\  =\ Linear\ Combination\ of\ Columns\ of\ A$

```julia
A = [1 4 5; 3 2 5; 2 1 3]
```

```julia
x_plane = Tuple(A[1,:])
y_plane = Tuple(A[2,:])
z_plane = Tuple(A[3,:])

surface([x_plane, y_plane, z_plane], fillcolor=:reds)
```
