Question 2

For vectors $v = \begin{equation}
\left[
\begin{array}{c}
4 \\
1 \\
\end{array}
\right]
\end{equation}$ and $w = \begin{equation}
\left[
\begin{array}{c}
-2 \\
2 \\
\end{array}
\right]
\end{equation}$ their graphical representations and combinations look like this:

```julia
plot([0, 4], 
     [0, 1],
     label = L"$v$",
     arrow = :head,
     linewidth = 5);
```

```julia
plot!([0, -2], 
     [0, 2],
     label = L"$w$",
     arrow = :head,
     linewidth = 5);
```

```julia
plot!([0, 2], 
     [0, 3],
     label = L"$v + w$",
     arrow = :head,
     linewidth = 5);
```

```julia
plot!([0, -6], 
     [0, 1],
     framestyle = :zerolines,
     title = "Graphical Representation",
     label = L"$v - w$",
     arrow = :head,
     linewidth = 5,
     legend = :best,
     xlim = (-8, 8),
     ylim = (-5, 5))
```
